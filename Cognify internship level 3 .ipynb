```python
from textblob import TextBlob
from collections import Counter



# Function to perform sentiment analysis and extract keywords
def analyze_rating_texts(rating_texts):
    positive_keywords = []
    negative_keywords = []

    for rating_text in rating_texts:
        blob = TextBlob(rating_text)
        sentiment_score = blob.sentiment.polarity

        # Classify as positive or negative based on sentiment score
        if sentiment_score > 0:
            positive_keywords.extend(blob.words)
        elif sentiment_score < 0:
            negative_keywords.extend(blob.words)

    # Count the occurrences of each word
    positive_word_counts = Counter(positive_keywords)
    negative_word_counts = Counter(negative_keywords)

    # Get the most common positive and negative keywords
    most_common_positive_keywords = positive_word_counts.most_common(10)
    most_common_negative_keywords = negative_word_counts.most_common(10)

    return most_common_positive_keywords, most_common_negative_keywords

# Analyze rating texts and extract keywords
positive_keywords, negative_keywords = analyze_rating_texts(df['Rating text'])

print("Most common positive keywords:")
print(positive_keywords)

print("\nMost common negative keywords:")
print(negative_keywords)
```

We define a function analyze_rating_texts() to calculate sentiment scores and extract keywords from the rating texts.
We iterate over each rating text, use TextBlob to calculate sentiment scores, and classify the text as positive or negative based on the sentiment score.
We extract words from positive and negative rating texts separately and count the occurrences of each word using Counter.
We retrieve the most common positive and negative keywords using the most_common() method.

```python
# Calculate the length of each review
df['Rating text'] = df['Rating text'].apply(lambda x: len(str(x)))

# Calculate the average length of reviews
average_review_length = df['Rating text'].mean()
print("Average Review Length:", average_review_length)

# Explore the relationship between review length and rating
relationship = df.groupby('Rating text')['Rating text'].mean()
print("\nRelationship between Review Length and Rating:")
print(relationship)
```

# Task: Votes Analysis

```python
# Find the restaurant with the highest number of votes
restaurant_highest_votes = df.loc[df['Votes'].idxmax()]

# Find the restaurant with the lowest number of votes
restaurant_lowest_votes = df.loc[df['Votes'].idxmin()]

print("Restaurant with the highest number of votes:")
print(restaurant_highest_votes)

print("\nRestaurant with the lowest number of votes:")
print(restaurant_lowest_votes)
```

# df['Votes'].idxmin()

# Task: Price Range vs. Online Delivery and Table Booking


```python
# Group the data by the combination of 'Price Range', 'Online Delivery', and 'Table Booking'
grouped = df.groupby(['Price range', 'Has Online delivery', 'Has Table booking']).size().reset_index(name='Frequency')

# Print the grouped data
print(grouped)
```

Grouping the Data:

The code takes your DataFrame df and groups it based on three columns: 'Price Range', 'Online Delivery', and 'Table Booking'.
This means it organizes your data into groups where all rows with the same combination of price range, online delivery availability, and table booking availability are grouped together.
Calculating the Frequency of Each Combination:

Once the data is grouped, the code calculates the size of each group.
This effectively counts how many rows belong to each combination of price range, online delivery availability, and table booking availability.
Resetting the Index and Naming the Frequency Column:

After calculating the sizes of the groups, the code resets the index of the resulting DataFrame.
This step converts the grouped data into a format where the combinations of price range, online delivery, and table booking become columns.
The 'Frequency' column is created to hold the counts calculated in the previous step.
Printing the Result:

Finally, the code prints the resulting DataFrame, which now shows each combination of price range, online delivery, and table booking, along with the frequency of occurrences for each combination.

```python
# Convert 'Online Delivery' and 'Table Booking' columns to boolean values
df['Online Delivery'] = df['Has Online delivery'].map({'Yes': True, 'No': False})
df['Table Booking'] = df['Has Table booking'].map({'Yes': True, 'No': False})

# Group the data by 'Price Range' and calculate the proportion of restaurants offering online delivery and table booking
service_proportion = df.groupby('Price range')[['Has Online delivery', 'Has Table booking']].mean()

print("Proportion of restaurants offering online delivery and table booking services by price range:")
print(service_proportion)
```
